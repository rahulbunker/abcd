```python
# lcm
x=6
y=10
def lcm(x,y):
    if x>y:
        big = x
    else :
        big =y

    while(True):
        if (big % x==0 and big %y ==0):
            print("lcm is :" , big)
            break
        
        big = big +1


lcm(3,5)
```

```python
# hcf

def hcf(x,y):
    if x>y:
        max = x
    else:
        max = y

    for i in range(2,max+1):
        if (x%i==0 and y%i==0):
            hcf = i
            break
         
            
    print("hcf is :" ,hcf)
hcf(10,15)
```

```python
def fact(n):
    fact=1
    for i in range(1,n+1):
        fact = fact *i
    print(fact)

fact(7)
        
```
